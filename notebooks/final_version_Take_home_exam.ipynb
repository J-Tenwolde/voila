# **Loading the dataset**

```python
# pandas module for working with Series and Dataframes
import pandas as pd
import altair as alt
#imprt altair & widgets
import altair as alt
import ipywidgets as widgets
from ipywidgets import interact, fixed
# load dataset
HealthFactorsData= pd.read_csv('https://raw.githubusercontent.com/NHameleers/dtz2025-datasets/master/CountryHealthFactors.csv')
#numbers of rows and columns 
HealthFactorsData.shape
```

```python
HealthFactorsData.columns
```

The result of the code above, shows that all the column headings are formulated differently. Some of the headings start with a capital while others do not. Besides this, the heading 'under-five deaths' contains an extra space after the text. Due to this space, the 'under-five deaths' leads to an error while using .loc for selecting the relevant columns for the research question. The problem was solved by renaming all column names with the same format. By using the **strip** function in combination with the **lower** function, all the headings starting with a capital letter will change into a lowercase letter and all the extra spaces will be removed. By doing this, the variables are easier to use in the data analysis.

```python
# issue of the names and solution
hf=HealthFactorsData.rename(columns=str.strip).rename(columns=str.lower)
hf
```

# **Describing and visualizing the data**

With the use of the **groupby** function, the mean, standard deviation and the quartile range of the mortality in the different age catagories and the total expenditure of each country are shown. 

```python
#This shows the mean, standard deviation and quartile range of the mortality in different age catagories and the total expenditure of each country
outcomes.groupby('country')['adult mortality', 'infant deaths', 'under-five deaths', 'total expenditure'].describe()
```

This table gives a clearer view of the mean mortality for the different age catagories and the mean total expanditure per country over 15 years



```python
#This shows the mean of mortality for each age catagorie, the total expenditure and the mean population over 15 years with the status column added
mean_outcomes = outcomes.groupby('country').agg(status=('status', 'first'),
                                mean_adult_mortality= ('adult mortality', 'mean'), 
                                mean_infant_deaths=('infant deaths', 'mean'), 
                                mean_underfive_deaths=('under-five deaths', 'mean'), 
                                mean_total_expenditure=('total expenditure', 'mean')).reset_index()
mean_outcomes
```

# **Interactive data description & visualization**

```python
#adjust visualize_mortality_data to take y-var as input
def visualize_mortality_data(y_var, mean_outcomes):
  
  chart = alt.Chart(mean_outcomes).mark_point().encode(
      x= 'mean_total_expenditure:Q', 
      y=f'mean({y_var}):Q', 
      column = 'status', 
      color = 'status',
      shape = 'status', 
      tooltip='country').properties(title = 'relationship between health expanditure and mortality')
  
  display(chart)



```

```python
# with this interaction the difference in adult mortality, under-five deaths and infant deaths can be compared between the developed and the undeveloped countries 
interact(visualize_mortality_data,
          y_var= widgets.Dropdown(options= ['mean_adult_mortality', 'mean_infant_deaths', 'mean_underfive_deaths']),
        
         mean_outcomes=fixed(mean_outcomes));

  
```
